```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def Re30(Re):
    '''
    compute Nusselt Number for Reynolds Number less than 30
    '''
    return 1.47 * np.power(Re, -1/3)
```

```python
def Re30_1800(Re):
    '''
    compute Nusselt Number for Reynolds Number between 30 and 1800
    '''
    up = Re
    down = (1.08 * np.power(Re,1.22)) - 5.2
    return up/down
```

```python
def Re1800(Re,Pr):
    '''
    compute Nusselt Number for Reynolds Number more than 1800
    '''
    up = Re
    down = 8750 + (58 * np.power(Pr,-0.5) * (np.power(Re, 0.75) - 253))
    return up/down
```

```python
def RE(Re, Pr):
    if Re <= 30:
        return Re30(Re)
    elif 30 < Re <= 1800:
        return Re30_1800(Re)
    else:
        return Re1800(Re,Pr)
```

```python
a = np.linspace(10,100,200)

b1 = Re30(a)
b2 = Re30_1800(a)

plt.figure(figsize=(18,10))

plt.plot(a,b1,label='formula 10-38, $Re_\delta$ < 30',linewidth=4)
plt.plot(a,b2,label='formula 10-39, 30 <$Re_\delta$ < 1800',linewidth=4)

plt.legend(fontsize=20)

plt.title('corrected Nusselt Number')

plt.xlabel('$Re_\delta$',fontsize=30)

plt.ylabel('$Nu$',fontsize=30)
_ = plt.xticks(fontsize=20)
_ = plt.yticks(fontsize=20)
```

```python
def res(array, Pr):
    b = []
    for i in array:
        b.append(RE(i,Pr))
    return np.array(b)
```

```python
a = np.linspace(10,10000,1000)

plt.figure(figsize=(22,12))

for i in range(1,10):
    b = res(a,i)
    plt.plot(a,b,label = '$Pr = {}$'.format(i),linewidth=4)

plt.legend(fontsize=20)   
plt.xlabel('$Re_\delta$',fontsize=30)
plt.title('corrected Nusselt Number', fontsize=50)
plt.ylabel('$Nu$',fontsize=30)

_ = plt.xticks(fontsize=20)
_ = plt.yticks(fontsize=20)
```
